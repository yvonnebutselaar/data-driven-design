```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
df = pd.merge(movies, ratings, on='movieId')
df.head(5)
```

```python
pt = pd.pivot_table(data=df, index='userId', columns='title', values='rating')
pt = pt.reindex(pt.count().sort_values(ascending=False).index, axis=1)
pt = pt.fillna(0)
pt.head(3)
```

I chose 

```python
corr = pt.corrwith(pt['Forrest Gump (1994)'])
corr = corr.sort_values(ascending=False)
```
